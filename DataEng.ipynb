```python
object_type_cols = [col for col in data.columns if pd.Series(data[col]).dtype == 'O']
for val in object_type_cols:
    print(val)
```

```python
columns_values = []
for feature in object_type_cols:
    column = [val for val in data[feature] if not pd.isnull(val)]
    column = list(set(column))
    columns_values.append(column)
dictionary = {}
index = 0
for feature in object_type_cols:
    dictionary[feature] = columns_values[index]
    index += 1
```
